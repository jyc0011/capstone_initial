```python
# 서버 프로그램 (TCP)
import socket, time
import random
masterNode = '192.168.0.10'
# 마스터 노드의 ip 주소. 새로운 연결 노드에 대한 정보를 받을때, 사용된다.(고정됨)
clientNode = '0.0.0.0'
# 연결될 클라이언트 ip 주소. 본 노드는 서버 노드가 되고, 연결될 노드는 클라이언트 노드가 되어서 연결된다. 
serverNode = '0.0.0.0'
# 연결될 서버 ip 주소, 본 노드는 클라이언트 노드가 되고, 연결될 노드는 서버 노드가 되어서 연결된다.
connectfullnode = []
# 연결될 두 노드에 대한 ip주소 저장

# 1. 마스터 노드와 풀노드간의 연결 
# 서버 프로그램은 주피터, 이것은 파이참으로 실행 - 동시에 두개 프로그램 실행을 위해서.
server_ip = masterNode # 위에서 설정한 서버 ip
server_port = 5000 # 마스터 노드의 포트번호

mastersocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
# 마스터 노드의 소켓에 연결
mastersocket.connect((server_ip, server_port))
# 마스터 노드의 소켓에 연결

sendmsg = input('msg:') # 서버로 보낼 msg 입력
mastersocket.sendall(sendmsg.encode(encoding='utf-8'))

# 서버가 에코로 되돌려 보낸 메시지를 클라이언트가 받음
data = socket.recv(100)
recemsg = data.decode() # 읽은 데이터 디코딩
print('echo msg:', recemsg)
if recemsg == "This is ServerNode ip " + *:
    serverNode = msg.replace('This is ServerNode ip ','')
    # 서버 노드의 ip주소 추출
    mastersocket.close()
    # 현재 마스터 노드와 연결되있는 소켓을 닫고 자신이 클라이언트노드로써 새로운 소켓에 연결하기
else if recemsg == "This is ClientNode ip " + *:
    clientNode = msg.replace('This is ServerNode ip ','')
    # 클라이언트 노드의 ip주소 추출
    mastersocket.close()
    # 현재 마스터 노드와 연결되있는 소켓을 닫고 자신이 서버노드로써 새로운 소켓에 연결하기
    iamserver(clientNode)
    # 자신을 서버노드로 클라이언트 노드와 연결하는 함수
socket.close()


def iamserver(clientnode):
    localhost = socket.gethostbyname(socket.gethostname())
    # 자신의 ip주소를 받아옴
    host =localhost # 서버 컴퓨터의 ip(여기선 내 컴퓨터를 서버 컴퓨터로 사용) 
                       # 본인의 ip주소 넣어도 됨(확인방법: cmd -> ipconfig)
    serverport = 3333  
    # 서버 포트번호(다른 프로그램이 사용하지 않는 포트번호로 지정해야 함)
    # 해당 포트번호를 블록체인 네트워크 상의 풀노드 서버 포트로 설정(우리는 풀노드간의 통신에서 서버 포트를 3333이라고 고정한다.)

    # 서버소켓 오픈(대문을 열어둠)
    # socket.AF_INET: 주소종류 지정(IP) / socket.SOCK_STREAM: 통신종류 지정(UDP, TCP)
    # SOCK_STREAM은 TCP를 쓰겠다는 의미
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    # 여러번 ip.port를 바인드하면 에러가 나므로, 이를 방지하기 위한 설정이 필요
    # socket.SOL_SOCKET: 소켓 옵션
    # SO_REUSEADDR 옵션은 현재 사용 중인 ip/포트번호를 재사용할 수 있다.
    # 커널이 소켓을 사용하는 중에도 계속해서 사용할 수 있다
    server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

    # server socket에 ip와 port를 붙여줌(바인드)
    server_socket.bind((host, serverport))
    # 클라이언트 접속 준비 완료
    server_socket.listen()
    print('echo server start') # echo program: 입력한 값을 메아리치는 기능(그대로 다시 보냄)
    # accept(): 클라이언트 접속 기다리며 대기
    # 클라이언트가 접속하면 서버-클라이언트 1:1 통신이 가능한 작은 소켓(client_soc)을 만들어서 반환
    # 접속한 클라이언트의 주소(addr) 역시 반환
    client_soc, addr = server_socket.accept()
    # 연결된 클라이언트 노드의 ip주소와 port(동적)번호를 받아옴
    print('connected client addr:', addr)
    if clientnode == addr[0]:
        # 만약 마스터 노드한테 받은 클라이언트 노드의 ip주소와 현재 연결된 클라이언트 주소가 일치할 경우
        # 해당 클라이언트와 연결하여 소켓 통신을 함.
        sendmsg_connectrequest = input("Connect BlockChain Node : Client")
        # 나와 연결된 노드가 클라이언트, 즉 자신은 서버노드이고 메세지를 전송받은 노드가 클라이언트로 연결하겠다는 확인 요청 메세지
        client_soc.sendall(sendmsg_connectrequest.encode(encoding='utf-8'))
        # 인코딩 작업을 거쳐 클라이언트 소켓으로 전송됨.

        data = client_soc.recv(100)
        recemsg_connect = data.decode() # 읽은 데이터 디코딩
        # 소켓에 읽을 데이터가 없으면 생길 때까지 대기함
        # recv(메시지크기): 소켓에서 크기만큼 읽는 함수
        print('recv msg:',recemsg)

        if recemsg_connect == "Okay. i am your Client. Can i Receive your Chain?":
            recemsg_connect
        
        data = client_soc.recv(100)
        recemsg = data.decode() # 읽은 데이터 디코딩
        print('recv msg:', recemsg)
        if msg == "This is ServerNode ip " + *:
            serverNode = msg.replace('This is ServerNode ip ','')
            # 서버 노드의 ip주소 추출
        else if msg == "This is ClientNode ip " + *:
            clientNode = msg.replace('This is ServerNode ip ','')
            # 클라이언트 노드의 ip주소 추출
        client_soc.sendall(msg.encode(encoding='utf-8')) # 에코메세지 클라이언트로 보냄

        time.sleep(5)
        server_socket.close() # 사용했던 서버 소켓을 닫아줌
    else:
        server_socket.close() # 사용했던 서버 소켓을 닫아줌
```

```python
# 클라이언트 프로그램
# 서버 프로그램은 주피터, 이것은 파이참으로 실행 - 동시에 두개 프로그램 실행을 위해서.
import socket

server_ip = '192.168.0.11' # 위에서 설정한 서버 ip
server_port = 3333 # 위에서 설정한 서버 포트번호

socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
socket.connect((server_ip, server_port))

msg = input('msg:') # 서버로 보낼 msg 입력
socket.sendall(msg.encode(encoding='utf-8'))

# 서버가 에코로 되돌려 보낸 메시지를 클라이언트가 받음
data = socket.recv(100)
msg = data.decode() # 읽은 데이터 디코딩
print('echo msg:', msg)

socket.close()
```
